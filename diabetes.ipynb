```python
from sklearn.linear_model import LinearRegression    #Simple rÃ©gression linÃ©aire
from sklearn.preprocessing import StandardScaler  #Scale les donnÃ©es

scaler = StandardScaler()


X_train, y_train = train_data_original.drop(['Outcome'], axis=1), train_data_original['Outcome']
X_train_s = scaler.fit_transform(X_train)

reg = LinearRegression()

reg.fit(X_train_s, y_train)
```

```python
X_test_s = scaler.transform(X_test)
reg.score(X_test_s, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(X_train_s, y_train)
```

```python
forest.score(X_test_s, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    "n_estimators": [100, 200, 300],
    "min_samples_split": [2, 4],
     "max_depth": [None, 4, 8]
    }

grid_search = GridSearchCV(forest, param_grid, cv=5, scoring="neg_mean_squared_error", return_train_score=True)

grid_search.fit(X_train_s, y_train)
```

```python
best_forest = grid_search.best_estimator_
best_forest.score(X_test_s, y_test)
```
